### Problem 1:
#### Problem Statement
Draw a pie plot based on number of wins for each team

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
class a():
    sizes=[]
    df=pd.read_csv('DataFiles/Match.csv')
    teams=np.unique(df.Team1)
    def wins(x,df):
        count=0
        for i in range(len(df)):
            if x == df.match_winner[i]:
                count=count+1
        return count
    for x in teams:
        z=wins(x,df)
        sizes.append(z)
    plt.pie(sizes,labels=teams,autopct='%1.1f%%',startangle=140)
    plt.show()
pie_plot=a
```

### Problem 2:
#### Problem Statement
Draw a bar graph for number of matches won (i,e.consider win if the team wins both the toss and match)  for the team “Mumbai Indians” for all years 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
class a():
    def __init__(self,teams):
        def matcheswon(y,teams,df):
            count=0
            for i in range(len(df)):
                if (y==df.Season_Year[i]) and teams==df.match_winner[i] and teams==df.Toss_Winner[i]:
                    count=count+1
            return count
        df=pd.read_csv('DataFiles/Match.csv')
        years=np.unique(df.Season_Year)
        y=[]
        for x in years:
            z=matcheswon(x,teams,df)
            y.append(z)
        width=1/3
        plt.bar(years,y,width,color='yellow')
bar=a('Mumbai Indians')
```
