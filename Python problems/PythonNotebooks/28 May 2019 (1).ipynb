### Problem 1:
#### Problem Statement
* Print no. of matches won by each team based on year wise (each year as one column)

```python
import numpy as np
import pandas as pd
df1=pd.read_csv('DataFiles/Match.csv',usecols=['Season_Year'])
years=np.unique(df1.values)
df2=pd.read_csv('DataFiles/Match.csv',usecols=['Team1'])
teams=np.unique(df2.values)
df5=pd.DataFrame()
df5['teams']=teams
df3=pd.read_csv('DataFiles/Match.csv',usecols=['match_winner'])
for x in years:
    a=[]
    for j in teams:
        count=0
        for i in range(len(df3)):
            if j==df3.values[i] and df1.values[i]==x:
                count=count+1
        a.append(count)
    df5[x]=a
df5
```

### Problem 2:
#### Problem Statement
Print  ManOfMach count of each player in  Hyderabad stadium

```python
import pandas as pd
import numpy as np
df=pd.read_csv('DataFiles/Match.csv')
df2=pd.read_csv('DataFiles/Match.csv',usecols=['Team1'])
teams=np.unique(df2.values)
players=[]
for i in range(len(df)):
    if df.values[i][14] not in teams:
        if df.values[i][6]=='Rajiv Gandhi International Stadium, Uppal':
            players.append(df.values[i][14])
player=np.unique(players)
number=[]
for x in player:
    count=0
    for i in range(len(players)):
        if x==players[i]:
            count=count+1
    number.append(count)
df2=pd.DataFrame()
df2['players']=player
df2['ManOfMatches']=number
df2
```
