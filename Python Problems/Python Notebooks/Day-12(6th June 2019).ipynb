### 1.Missing Values

### 2.Categorical Data

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df = df.fillna(method = "ffill")
dfle = df
dfle.Country = le.fit_transform(df.Country)
dfle.Purchased = le.fit_transform(df.Purchased)
dfle
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(categorical_features = [0])
x = ohe.fit_transform(df).toarray()
x = pd.DataFrame(x)
x
```

### 3.Splitting Data
